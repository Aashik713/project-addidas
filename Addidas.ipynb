```python
for i in range(9648):
    ad['Operating Profit'][i]=ad['Operating Profit'][i].replace('$',"")
    ad['Operating Profit'][i]=ad['Operating Profit'][i].replace(',',"")
    ad['Operating Profit'][i]=ad['Operating Profit'][i].replace(' ',"")
    ad['Price per Unit'][i]=ad['Price per Unit'][i].replace('$',"")
    ad['Price per Unit'][i]=ad['Price per Unit'][i].replace(',',"")
    ad['Price per Unit'][i]=ad['Price per Unit'][i].replace(' ',"")
    ad['Price per Unit'][i]=ad['Price per Unit'][i].replace('.',"")
    ad['Total Sales'][i]=ad['Total Sales'][i].replace('$',"")
    ad['Total Sales'][i]=ad['Total Sales'][i].replace(',',"")
    ad['Total Sales'][i]=ad['Total Sales'][i].replace(' ',"")
```

```python
ad['Price per Unit']=pd.to_numeric(ad["Price per Unit"])
ad['Total Sales']=pd.to_numeric(ad["Total Sales"])
ad['Operating Profit']=pd.to_numeric(ad["Operating Profit"])
for i in range (9648):
    ad['Price per Unit'][i]=ad['Price per Unit'][i]/100
    
    #ad['Operating Profit'][i]=int(ad['Operating Profit'][i]/10)
    #ad['Total Sales'][i]=int(ad['Total Sales'][i]/10)
```

```python
#total unit sold
p=ad['Product'].unique()
s=0
e=[]
for i in range(len(p)):
    for j in range(9648):
        if(ad['Product'][j]==p[i]):
            s=s+ad['Units Sold'][j]
    e.append(s)
    print(p[i]," - ",s)
    s=0
```

```python
#total sales
p=ad['Product'].unique()
s=0
ps=[]
for i in range(len(p)):
    for j in range(9648):
        if(ad['Product'][j]==p[i]):
            s=s+ad['Total Sales'][j]
    ps.append(s)
    print(p[i]," - ",int(s))
    s=0
```

```python
Re=ad['Region'].unique()
s=0
Res=[]
for i in range(len(Re)):
    for j in range(9648):
        if(ad['Region'][j]==Re[i]):
            s=s+ad['Total Sales'][j]
    Res.append(s)
    print(Re[i]," - ",int(s))
    s=0
```

```python
R=ad['Retailer'].unique()
s=0
Rs=[]
for i in range(len(R)):
    for j in range(9648):
        if(ad['Retailer'][j]==R[i]):
            s=s+ad['Total Sales'][j]
    Rs.append(s)
    print(R[i]," - ",int(s))
    s=0
```

```python
Sm=ad['Sales Method'].unique()
s=0
k=0
Sms=[]
for i in range(len(Sm)):
    for j in range(9648):
        if(ad['Sales Method'][j]==Sm[i]):
            s=s+ad['Units Sold'][j]
            k=k+ad['Total Sales'][j]
    Sms.append(k)
    print(Sm[i]," - ",s," => ",k)
    s=0
    k=0
```

```python
S=ad['State'].unique()
s=0
c=[]
for i in range(len(S)):
    for j in range(9648):
        if(ad['State'][j]==S[i]):
            s=s+ad['Total Sales'][j]
    print(S[i]," - ",int(s))
    s=0
```

```python
C=ad['City'].unique()
s=0
Cs=[]
for i in range(len(C)):
    for j in range(9648):
        if(ad['City'][j]==C[i]):
            s=s+ad['Total Sales'][j]
    Cs.append(s)
    print(C[i]," - ",s)
    s=0
```

```python
p=ad['Product'].unique()
s=0
l=0
for i in range(len(p)):
    for j in range(9648):
        if(ad['Product'][j]==p[i]):
            s=s+ad['Price per Unit'][j]
            l+=1
    print("average cost per item ",p[i]," - ",int(s/l))
    s=0
    l=0
```

```python
#average prepare cost per item
p=ad['Product'].unique()
s=0
for i in range(len(p)):
    for j in range(9648):
        if(ad['Product'][j]==p[i]):
            s=s+ad['Total Sales'][j]-ad['Operating Profit'][j]
            l+=1
    s=s/e[i]
    print(p[i]," - ",int(s))
    s=0
```

```python
# average profit cost per item
p=ad['Product'].unique()
s=0
l=0
for i in range(len(p)):
    for j in range(9648):
        if(ad['Product'][j]==p[i]):
            s=s+ad['Operating Profit'][j]
    s=s/e[i]
    print("average operating profit ",p[i]," - ",int(s))
    s=0
```
